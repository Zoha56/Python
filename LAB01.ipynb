```python
#Q2 – Count Total Patients Count how many patients exist in the dataset. 
def count_patients(converted):
   return len(converted) #since patients are non-repeating
print("count of patients is: ",count_patients(converted))
```

```python
#Q3 – Average Heart Rate Per Doctor Calculate average heart rate grouped by doctor.
def average_Hr(converted):
    average={} #to store the average per doctor
    doctor_count={} #how many time doctor appears (in short patient count for each doctor)
    hr_sum={} #sum of patients heart rate per doctor
    for patients in converted:
        doctor=patients['doctor']
        if not doctor in doctor_count: #if doctor not in doctor_count add it in that dictionary
            doctor_count[doctor]=1
            hr_sum[doctor]=patients['hr']
        else: 
            doctor_count[doctor]+=1 #else count+1
            hr_sum[doctor]+=patients['hr']
    for doctor in doctor_count:
        average[doctor]=hr_sum[doctor]/doctor_count[doctor]
    return average
avg=average_Hr(converted)
for k,v in avg.items():
    print(k,": ",v)
```

```python
#Q4 – High-Risk Patients (HR > 100) List all patients whose heart rate is above 100.
def high_rate(converted):
    patients=[]
    for patient in converted:
        if patient['hr']>100: #if patient heart rate >100 add his/her name in the list
            patients.append(patient['name'])
    return patients
high_risk=high_rate(converted)
print("patients with high risk ")
for p in high_risk:
    print(p)
```

```python
#Q5 – Lowest Glucose Level Find the patient having the lowest glucose reading.
def low_glucose(converted):
    lowest=200 #starting with highest possible value of glucose
    patientname=None
    for patient in converted:
        if patient['glucose']<lowest:  #if patient glucose is less than current value stored in lowest , change lowest to that value
            lowest=patient['glucose']
            patientname=patient['name']
    return patientname
print("lowest glucose: ",low_glucose(converted))
```

```python
#Q8 – Distinct Doctors Count Count the total number of unique doctors.
def distinct_count(converted):
    doctors=list() #to store name of doctors
    count=0 #to store count of each  unique doctor
    for doctor in converted:
        unique=doctor['doctor']
        if not unique in doctors:
             doctors.append(unique)
             count+=1
    return count
print("distinct doctors:",distinct_count(converted))
```

```python
#Q9 – Feedback Distribution: Count how many patients fall under each feedback type (Excellent, Good, Fair, Critical). 
def feedback(converted):
    feedback_count={} #to store count of each feedback along with its name
    for patient in converted:
        review=patient['feedback']
        if not review in feedback_count: #if not in dictornary add it and value=1
            feedback_count[review]=1
        else:
            feedback_count[review]+=1 #else value=+1
    return feedback_count
feedback_count=feedback(converted)
for fd,count  in feedback_count.items():
        print(fd,':',count)
```

```python
#Q10 – Overall Average Heart Rate Calculate average HR across all patients. 
def average_all_hr(converted):
    sum_hr=0 #sum of all heart rates
    count=count_patients(converted) #total patients
    for patient in converted:
        sum_hr+=patient['hr']
    avg=sum_hr/count
    return avg
print("Average heart rate: ",average_all_hr(converted))
```

```python
#Q11 – Health Grade AssignmentAssign health grade: 
#• HR < 70 = A (Excellent) 
#• HR 70–89 = B (Stable) 
#• HR 90–110 = C (Concern) 
#• HR > 110 = D (High Risk) 
def hr_grade(converted):
    for data in converted: #assignning the health grade according to given criteria
        if data['hr']<70:
            data['HeatlhGrade']='A'
        elif data['hr'] in range(70,90):
            data['HeatlhGrade']='B'
        elif data['hr']in range(90,111):
            data['HeatlhGrade']='C'
        elif data['hr']>110:
            data['HeatlhGrade']='D'
    return converted
grading=hr_grade(converted)
for patients in grading:
    print(patients)
```

```python
#Q12 – Glucose Improvement ( − 5%) Assume glucose is reduced by 5% for every patient.
def updated_glucose(converted):
    for data in converted:
        glucose=data['glucose']
        new_glucose=glucose*0.95 #new glucose will be 95% of previous glucose
        data['glucose']=int(new_glucose)
    return converted
updated=updated_glucose(converted)
for glucose in updated:
    print(glucose)
```

```python
#Q13 – Doctor Summary For each doctor show: 
#• number of assigned patients 
#• average HR 
#• average glucose 
def doctor_summary(converted):
    doctors={} #store doctors name along with patient count, averge hr and glucose
    for data in converted:
        doctor=data['doctor']
        hr=data['hr']
        glucose=data['glucose']
        if not doctor in doctors:
            doctors[doctor]={  #if doctor not in dictionary intialize values
                'sum_hr':0,
                'sum_glucose':0,
                'patient count':0
                }
        doctors[doctor]['sum_hr']+=hr #assign values of current record
        doctors[doctor]['sum_glucose']+=glucose
        doctors[doctor]['patient count']+=1
    for doctor in doctors:
        doctors[doctor]['avg hr']=doctors[doctor]['sum_hr']/doctors[doctor]['patient count']
        doctors[doctor]['avg glucose']=doctors[doctor]['sum_glucose']/doctors[doctor]['patient count']
        del doctors[doctor]['sum_hr'] #delete extra keys
        del doctors[doctor]['sum_glucose']
    return doctors
doctors=doctor_summary(converted)
for doctor,summary in doctors.items():
  print(doctor)
  for k,v in summary.items():
      print(k,':',v)
```

```python
#Q15 – Critical Condition Tag Mark patients with either: 
#• BP > 140/90 
#• OR feedback = "Critical" 
#As "Severe", else "Normal"
def patient_condition(converted):
    for data in converted:
        if data['bp']>"140/90" or data['feedback']=='Critical': #specifying the condition criteria 
            data['condition']='Severe'
        else:
            data['condition']='Normal'
    return converted
condition=patient_condition(converted)
for p in condition:
    print(p)
```

```python
#Q17 – Underrated Cases Patients having normal HR (<90) but feedback “Fair” or “Critical”.
def underrated(converted):
    underrated_cases=[]
    for data in converted:
        if data['hr']<90 and (data['feedback']=='Fair' or data['feedback']=='Critical'):
            underrated_cases.append(data['name'])
    return underrated_cases
print("Underrated cases:")
patients=underrated(converted)
for patient in patients:
      print(patient)
```

```python
#Q18 – Lowercase Feedback + Count 'critical' Convert all feedback to lowercase and count how many are “critical”.
def criticalcount(converted):
    count=0
    for data in converted:
        data['feedback']=data['feedback'].lower()
        if data['feedback']=='critical':
            count+=1
    return count
print("critical count:",criticalcount(converted))
```

```python
#Q19 – Doctor Load Check ,Verify each doctor has at least 2 patients. Return list of overworked or understaffed doctors.
def load(converted):
    doctor_count={}
    for data in converted:
        doctor=data['doctor']
        if not doctor in doctor_count:
            doctor_count[doctor]=1
        else:
            doctor_count[doctor]+=1
    loadwork=[]
    for d,count in doctor_count.items():
        if count<2 or count>10: #if doctor has less than 2 he/she is underwork and if greater than 10 he/she is overworked so  add them in list
            loadwork.append(d)
    return loadwork
work=load(converted)
for doctor in work:
     print(doctor)
```
