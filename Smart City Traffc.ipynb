```python
#Q1.Convert each traffic log into structured Python data with validation.
traffic_logs = [
"id:501,zone:A1,vehicle:Car,speed:62,time:08:30,violations:[None],status:Smooth",
"id:502,zone:A1,vehicle:Bike,speed:85,time:09:10,violations:[Helmet],status:Busy",
"id:503,zone:B2,vehicle:Bus,speed:45,time:17:25,violations:[None],status:Smooth",
"id:504,zone:C3,vehicle:Car,speed:110,time:14:15,violations:[Overspeed],status:Congested",
"id:505,zone:A1,vehicle:Truck,speed:40,time:18:50,violations:[None],status:Smooth",
]
def convert(traffic_logs):
    result=[]
    for data in traffic_logs:
        items=data.split(',')
        traffic={}
        for item in items:
            key,value=item.split(':',1) #split only at first
            key=key.strip() #to remove any extra spaces
            value=value.strip()
            if key=="id" or key=="speed":
                  value=int(value) #convert into integer for furture calculation
            elif key=="violations":
                value=value.strip("[]") #remove [] from start and end only
                value=value.split()#split values at blank spaces
                value=[v.strip().strip(',') for v in value]#
            traffic[key]=value
        result.append(traffic)
    return result
result=convert(traffic_logs)
for data in result:
   print(data)
```

```python
#Q2: Calculate average speed per zone.
def speed_per_zone(result):
   zone_speed={}#to store sum of speeds per zone
   zone_count={}#to count number of  zones
   for data in result:
       zone=data['zone']
       speed=data['speed']
       if not zone in zone_speed: 
          zone_speed[zone]=speed
          zone_count[zone]=1
       else:
           zone_speed[zone]+=speed
           zone_count[zone]+=1
   average={}
   for zone in zone_speed:
         average[zone]=zone_speed[zone]/zone_count[zone] #calculate average per zone
   return average
average_Speedperzone=speed_per_zone(result)
for zone,item in average_Speedperzone.items():
        print(zone,item)
```

```python
#Q3: Determine peak hour (hour with highest traffic entries).
def peak_hour(result):
    hour_count={}
    for data in result:
        time=data['time']
        hour=time.split(':')[0]
        if not hour in hour_count:
            hour_count[hour]=1
        else:
            hour_count[hour]+=1
    peak=max(hour_count,key=hour_count.get) #if two are max then return the first maximum
    count=hour_count[peak] #count of vehicals in that hour
    return peak,count
output=peak_hour(result)
print("hour: ",output[0])
print("count: ",output[1])
```

```python
#Q4: Find vehicles with speed > 80 km/h.
def fast_speed(result):
    fastspeed=[]#to store name of vehicles with speed>80 km/h
    for data in result:
        speed=data['speed']
        if speed>80:
            fastspeed.append(data['vehicle'])
    return fastspeed
name=fast_speed(result)
for v in name:
    print(v)
```

```python
#Q5: Count occurrences of each violation type.
def violationoccurance(result):
    violation_count={}#store violation name as key and count of violation as value
    for data in result:
        violation=data['violations']
        for v in violation:
           if v=='None': #skip if no violation is occurred
               continue
           elif not v in violation_count:
               violation_count[v]=1
           else:
              violation_count[v]+=1
    return violation_count
violationcount=violationoccurance(result)
for v,c in violationcount.items():
    print(v,c)
```

```python
#Q6: Compute safety index for each zone.
def compute_safety_index(result):
    zone_score = {}

    for data in result:
        zone = data['zone']
        speed = data['speed']
        violations = data['violations']
        status = data['status']
        risk = 0 #assigning points for each risk
        # speed risk
        if speed > 80:
            risk += 2
        # violation risk
        if violations[0] != "None":
            risk += 2
        if status == "Busy": # status risk
            risk += 1
        elif status == "Congested":
            risk += 3
        if zone not in zone_score:
            zone_score[zone] = 0
        zone_score[zone] += risk
    safety_index = {z: max(0, 10 - score) for z, score in zone_score.items()} #convert to safety index (lower risk â†’ higher index) to keep it positive
    return safety_index
index=compute_safety_index(result)
for zone,safe in index.items():
       print(zone,": ",safe)
```

```python
#Q7: Create a summary for each vehicle category.
def summary(result):
    summary={} #dictionary of vehicles
    for data in result:
        vehicle=data['vehicle']
        speed=data['speed']
        violation=data['violations']
        status=data['status']
        if vehicle not in summary: #if vehicle name is not in summary first make its dictionary member
            summary[vehicle]={
                'count':0,
                'totalspeed':0,
                'violations':[],
                'status':[]
            }
        summary[vehicle]['count']+=1
        summary[vehicle]['totalspeed']+=speed
        summary[vehicle]['violations'].extend(violation)
        summary[vehicle]['status'].append(status)
    for vehicle in summary:
        summary[vehicle]['avg_speed']=summary[vehicle]['totalspeed']/summary[vehicle]['count']
        del summary[vehicle]['totalspeed'] #deleting the extra key
    return summary
summary=summary(result)
for vehicle,data in summary.items():
    print(f"Vehical: {vehicle}")
    print(f"Count: {data['count']}")
    print(f"average speed: {data['avg_speed']}")
```

```python
#Q8: Identify high-congestion zones.
def high_cong(result):
    congest=set()#to keep zones unique
    for data in result:
        status=data['status']
        if status=="Congested":
            zone=data['zone']
            congest.add(zone)
    return congest
cong=high_cong(result)
for c in cong:
   print(c)
```

```python
#Q9: Classify each log into time windows (Morning, Afternoon, Evening, Night).
def define_window(result):
    for data in result:
        time=data['time']
        if time>="06:00" and time<="11:59": #defining the limit for each time and adding a new key to each dictionary of resultant list
            data['window']="morning"
        elif time>="12:00" and time<="16:59":
            data['window']="afternoon"
        elif time>="17:00" and time<="20:59":
            data['window']="evening"
        elif time>="21:00" and time<="05:59":
            data['window']="night"
    return result
updated=define_window(result)
for item in updated:
    print(item)
```

```python
#Q10: Generate final zone-level report (vehicles, avg speed, violations, common vehicle type, safety,category).
def zone_report(result):
    zone_summary = {}
    for data in result:
        zone = data['zone']
        speed = data['speed']
        vehicle = data['vehicle']
        violations = data['violations']
        if zone not in zone_summary:
            zone_summary[zone] = {
                "vehicles": [],
                "total_speed": 0,
                "count": 0,
                "violations": {},
            }

        
        zone_summary[zone]["vehicles"].append(vehicle)             # store vehicle
        zone_summary[zone]["total_speed"] += speed              # count speed for avg
        zone_summary[zone]["count"] += 1
        for v in violations:                          # count violations
            if v not in zone_summary[zone]["violations"]:
                zone_summary[zone]["violations"][v] = 1
            else:
                zone_summary[zone]["violations"][v] += 1

   # compute final formatted report
    final = {}

    for zone, info in zone_summary.items():
        avg_speed = info["total_speed"] / info["count"]

        from collections import Counter
        common_vehicle = Counter(info["vehicles"]).most_common(1)[0][0]         # Most common vehicle

        total_viol = sum(info["violations"].values())              # More violations = lower safety
        if total_viol == 0:
            safety = "safe"
        elif total_viol <= 2:
            safety = "medium"
        else:
            safety = "high risk"

        final[zone] = {
            "vehicles": info["vehicles"],
            "avg_speed": avg_speed,
            "violations": info["violations"],
            "common_vehicle": common_vehicle,
            "safety": safety
        }

    return final
final_summary=zone_report(result)
for k,v in final_summary.items():
    print(f"Zone:{k}")
    print(f"Vehicles:{v['vehicles']}")
    print(f"Average speed:{v['avg_speed']}")
    print(f"Violations:{v['violations']}")
    print(f"Common Vehicle:{v['common_vehicle']}")
    print(f"Safety:{v['safety']}")
    print(" ")
```
