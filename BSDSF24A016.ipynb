## Problem 5 – Friendship Network Analysis 
You are designing a social platform feature to suggest friends. Find mutual friends, unique
connections, and friend suggestions between two or more users.
#### Approach:
we have a dictionary of users along with the list of their friends. To find mutual and unique friends we treat users friends as sets and take their intersection and symmetric difference respectively. To find suggestions first combine all users then remove the user for which suggestions are to be found and then remove suggested friends from the set which are already user's friends.

```python
users={"A":["B","C","D"] , "B":["C","E","D"] , "C":["A","D","F"]}
def mutual(user1,user2):
    return set(users[user1])&set(users[user2])
def unique(user1,user2):
    return set(users[user1])^set(users[user2])
def suggestion(user):
    s=set()
    for friend in users:
        if friend in users:
            s.update(users[friend])
    s.discard(user)
    s -= set(users[user])
    return s

print("Mutual friends of A and B:", mutual("A", "B"))
print("Unique connections of A and B:", unique("A", "B"))
print("Friend suggestions for A:", suggestion("A"))
```

## Problem 15 – Expense & Budget Tracker
Create a financial tracker for daily expenses. Categorize expenses, calculate total per
category, check if spending exceeds budget, and find largest expense
#### Approach:
make list all daily expenses with their category, amount, and description. Set a budget for each category. Then, add up the amounts for each category to get the total spending per category. Compare these totals with the budget to see if any category exceeds its limit. then find the largest expense by checking which entry has the highest amount. 

```python

expenses = [
    {"category": "Food", "spent": 1000, "description": "Lunch"},
    {"category": "Fuel", "spent": 2000, "description": "Bus fare"},
    {"category": "Outing", "spent": 5000, "description": "Movie"},
    {"category": "Food", "spent": 1000, "description": "Dinner"},
    {"category": "Bills", "spent": 10000, "description": "Electricity"}]

budget = {"Food": 3000,
    "Fuel": 2000,
    "Outing": 2000,
    "Bills": 15000}

totals = {}
for e in expenses:
    categ = e["category"]
    if categ in totals:
        totals[categ] += e["spent"]
    else:
        totals[categ] = e["spent"]

for categ in totals:
    if totals[categ] > budget.get(categ, 0):
        print(f"Spending exceeded in {categ}  Spent: {totals[categ]}, Budget: {budget[categ]}")
largest = expenses[0]
for e in expenses:
    if e["spent"] > largest["spent"]:
        largest = e

print("Total Spending per Category:")
for category, total in totals.items():
    print(f"{category}: {total}")

print(f"largest Expense: {largest['description']} in {largest['category']} of {largest['spent']}")
```

## Problem 16 – Integrated Real-World System: Smart Business Manager 
Develop a 'Smart Business Manager' that combines all data-handling skills.

• Record business transactions with details like date, type, amount, and category
• Maintain all transactions for reporting

• Summarize totals by category
• Identify the highest expense
• Generate a formatted financial summary report
• Use modular functions such as add_transaction(), total_summary(), highest_expense(), etc.
